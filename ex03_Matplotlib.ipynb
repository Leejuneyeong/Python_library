```python
#라이브러리 불러오기
import matplotlib.pyplot as plt
import numpy as np
```

```python
#line Chart 그리기!
y = [2,4,6,8]

plt.plot(y)
# 데이터를 하나만 넣어주면 y축 데이터가 되고 x축은 인덱스가 자동부여된다

plt.show() # 출력할려면 plt.show()를 써주자
```

```python
# line Chart 그리기 2
x = [1,2,3,4]
y = [2,4,6,8]

plt.plot(x,y,ls = '--', lw=5, color='m', marker='o', mfc='y')

# Line Chart 속성값 변경하기!

#1. 라인의 스타일 설정가능!  - ls (Line Style) -> 문자열 형태
#2. 라인의 두께 지정하기 - lw (Line Width) -> 그냥 숫자형태로 적어주면 된다
#3. 라인의 색깔 변경하기 - color -> 선색과 마커의 색까지 변경
#4. 값의 위치를 표시 가능 - marker
#5. 마커의 색 지정 - mfc (marker face color)

plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,6,9,12,15]

plt.plot(x,y,ls = ':', lw=3, color='red', marker='o', mfc='y')
```

```python
# Line Chart 실습
x=np.arange(8)
y=[5,2,3,6,8,7,5,7]
plt.plot(x,y,ls = ':', lw=2, color='g', marker='o', mfc='g')
plt.plot(x,y,ls = ':', color='r', marker='d', mfc='b')
```

```python
x=np.arange(8)
y=[5,2,3,6,8,7,5,7]
plt.plot(x,y,ls="-.", lw='3', color='y',marker='*', mfc='w',ms='30', mew=2, mec='c')
# 그리도 : 격자눈금, 모눈종이
plt.grid()

# 그래프 숫자 표시 범위지정 (xlim,ylim)  - 축소 // 확대
#plt.xlim(-2,10) # (시작값, 끝값)
#plt.ylim(-5,12) 
plt.xlim(1,5)
plt.xlim(4,9)
# 반드시 plt.show() 안에 모든 속성을 써야하고 plt.show()는 마지막에 써야한다
plt.show()
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y,ls ='-.' , c = '#C79FEF', lw = 5, marker = '$::):$', ms = 20, mfc = 'w', mec = '#FF81C0',mew = 2)
plt.show()
```

```python
# marker='$()$' <- 마커모양을 맘대로 바꿀 수 있음
```

 ### 한개의 차트판에 여러개의 그래프 그리기!

```python
x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9,12]

plt.plot(x, label='x')
plt.plot(y, label='y')
plt.plot(z, label='z')

#범례 (참고사항) : 데이터의 종류를 표시할 때 사용한다.
plt.legend()

# 차트 제목 설정하기
plt.title('chart', loc='right')

# x축 데이터 이름 설정하기
plt.xlabel('x_axis')
# y축 데이터 이름 설정하기
plt.ylabel('y축')
plt.show()
```

### 요일별 교통사고 사망자 수 시각화하기!

```python
#csv 파일 읽어오기!
import pandas as pd
data=pd.read_csv('Traffic_Accident_2017.csv', encoding='euc-kr')
data
```

```python
#데이터 확인하기
data.info()
```

```python
# 데이터의 상위행 몇개만 가져와서 출력하기
data.head() # default 값은 5 # 속성값 확인 단축키 : shift+tab
```

```python
# data에서 요일만 가져오기
data['요일']
```

```python
# 각각 요일의 value 개수 세보기! -> value_counts()
data2=data['요일'].value_counts()
data2
```

```python
#월화수목금토일 순으로 데이터 가져오기!
y = data2[['월','화','수','목','금','토','일']]
y
```

```python
x=y.index
```

```python
# bar그래프 그려보기
#plt.bar(y.index,y) #(x축, y축)
plt.bar(x,y)
plt.show()
```

```python
# 차트 제목 설정하기
plt.bar(x,y)
plt.title('요일별 사망교통사고 건수', loc='center')

# x축 데이터 이름 설정하기
plt.xlabel('요일')
# y축 데이터 이름 설정하기
plt.ylabel('사고건수').set_rotation(0)
# y축 라벨 글씨 돌리는 방법 -> .set_rotation(몇도)

plt.show()
```

### 파이차트 그려보기!

```python
# 범주형 데이터를 시각화할 때 사용한다!
fruit = ['사과','레몬','감','수박']
ratio = [40.2, 15, 19.8, 25] #비율로 계산해서 출력해준다
explode=[0.1,0,0,0]
plt.pie(ratio,labels=fruit, autopct='%.2f%%',explode=explode, shadow=True)
# 그래프에서 표시해주는 방법 및 explode 떨어짐 부분과 그림자 shadow 주는법
plt.show()
```
